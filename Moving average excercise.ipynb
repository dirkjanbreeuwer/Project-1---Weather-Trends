```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import datetime
```

```python
df = pd.read_csv("moving-average-exercise.csv", thousands=",")
```

```python
# Converting the first column to datetime
df['Date'] = pd.to_datetime(df['Date'])
# Create a column with the week number of the year
df['Week'] = df['Date'].dt.week
# Create a rolling average column
df['Sales_7_MA'] = df['Sales'].rolling(window=7).mean()
```

```python
df.info()
```

```python
pl1 = sns.distplot(df.Sales)
```

```python
#Daily sales data
pl2 = sns.pointplot(data=df,x="Date",y="Sales")
```

```python
#Daily sales data
pl3 = sns.pointplot(data=df,x="Date",y="Sales_7_MA")
```

```python
#Weekly sales data
df_wk = df.groupby('Week').mean()
pl4 = sns.pointplot(data=df_wk,x=df_wk.index,y="Sales")
```
